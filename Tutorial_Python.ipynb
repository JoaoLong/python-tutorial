### Variáveis

O Python tem 5 tipos fundamentais de variáveis:

- Números: inteiros, float, long, complex.

```python
#Números 
n = 1
x = 2.1
z = 1+2j
z1 = 3e-5j

print(n)
print(x)
print(z)
print(x+n)
```

- Strings: criadas com ' ou ", ou até ''' para multi-line strings.

```python
#Strings
name = "John"

print(name)
print(name[0])
print(name + "Smith")
```

- Listas: arrays com entradas de vários tipos possíveis, escrevem-se com parêntesis retos.

```python
#Listas
list = ["John", 22, 71.4]

print(list)
print(list[0])
print(list + ["Smith", 23])

list[2] = 11;
print(list)

del list[2]
print(list)
```

- Tuplos: listas só de leitura, escrevem-se com parêntesis.

```python
#Tuplos
tuple = ("John", 22, 71.4)

print(tuple)
print(tuple[0])
print(tuple + ("Smith", 23))

#Código inválido: tuple[2] = 100
```

- Dicionários: a cada chave corresponde uma entrada, como um mapa em C++.

```python
#Dicionários
dict = {"name" : "John", "grade" : 16}
print(dict)
print(dict["name"])
dict["age"] = 20;
print(dict)
```

Para realizar a conversão entre os vários tipos de variáveis, *casting*, usam-se as funções seguintes:

```python
#Casting
y = 2
y1 = float(y)
print(y1)
s = chr(y)
print(s + "hello")
```

Os operadores em Python são os usuais, com a diferença de que as potências se escrevem com um asterisco duplo, ** . As variáveis booleanas são também as usuais.

```python
#Operadores e booleanos
print(2**4)
print(3>2 and 4>1)
print(True and False)
print(True or False)
print(not False)
```

Outros operadores úteis são:
- Pertença, *in*, para strings, listas, tuplos e dicionários.
- Identidade, *is*, que verifica se os endereços de memória são os mesmos.

```python
#Operadores identidade e de pertença
list = ["John", 22, 71.4]
tuple = ("John", 22, 71.4)

print(list is tuple)
print("John" in list)
```

### Controlo

Os blocos de código de controlo são também os já conhecidos: loops *for* e *while* e *if-else*.


- Loop *for*:

```python
#Loop for
for i in range(1,10) :
    print(i)

for i in "Python" :
    print (i)

list = ["John", 22, 71.4]
for i in list :
    print (i)      
```

- Loop *while*:

```python
#Loop while
count = 0
while (count < 9):
    print(count)
    count = count + 1

count = 0
while (count < 9):
    if (count == 6) :
        break
    print(count)
    count = count + 1   

count = 0
while (count < 9):
    count = count + 1  
    if (count == 6) :
        continue 
    print(count)
```

- Blocos *if-else*:

```python
#If-else
if ("John" in list):
    print("Correct")
else:
    print("Wrong")

if (x == 2):
    print ("Two")
elif (x == 3):
    print ("Three")
else :            
    print ("Four")
```

### Funções

As funções são uma das características mais importantes em programação. Em Python, declaram-se com a palavra-chave *def*, e tal como nas variáveis, não é necessário especificar o tipo de *return*. Veja-se que em Python, todos os argumentos são passados por referência.

```python
#Funções
def function(arg):
   return arg

print(function(2))

def func(string):
   print(string)
  
#Escreve "None"   
print(func("Hello world"))
```

Quando a função não retorna nada, o seu return é *None*, pelo que *print(function)* escreverá *None* na consola. Se for necessário fazer *print* da função, pode-se dar a volta ao código de outra forma.

```python
def func(string):
    return string
   
print(func("Hello world"))
```

Outro conceito útil é o de *lambda functions*, que se declaram numa linha, da seguinte forma:

```python
#Lambda functions
sum = lambda arg1, arg2: arg1 + arg2
print(sum(2,3))

mult = lambda arg1, arg2: arg1*arg2
print(mult(2,3))
```

### Ficheiros

Os ficheiros em Python também funcionam de forma semelhante a outras linguagens. Os modos de abertura são:

- Leitura: *r*
- Escrita: *w*
- Leitura e escrita: *r+* se o ficheiro já existir, *w+* se não (ou para fazer *overwrite* do ficheiro já existente)

A função *read()* toma como argumento o número de caracteres a ser lidos.

```python
#Ficheiros
f = open("foo.txt", "w")
print ("Name of the file: ", f.name)
f.write("Hello world")
f.close()

f1 = open("foo.txt", "r")
str = f1.read(2)
f1.close()
print(str)
```

### Classes

O Python suporta programação orientada a objetos, com todas as suas nuances. Deixa-se como exemplo um caso simples. 

O construtor é *init*, o destrutor é *del*, e dentro da classe definem-se as variáveis e métodos pretendidos. A palavra-chave *self* funciona como o *this* em C++. As classes em Python possuem uma documentação opcional, acessível por *doc*.

```python
#Classes e objetos
class Employee:
    #Documentation string - __doc__
    'Common base class for all employees'

    #Variáveis
    empCount = 0

    #Construtor
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
        Employee.empCount += 1

    #Destrutor
    def __del__(self):
        class_name = self.__class__.__name__
        print(class_name, "destroyed")
    
    #Métodos
    def displayCount(self):
        print("Total Employee ", Employee.empCount)

    def displayEmployee(self):
        print("Name : ", self.name,  ", Salary: ", self.salary)

print(Employee.__doc__)
emp1 = Employee("Zara", 2000)
```

As classes derivadas definem-se da seguinte forma:

```python
#Classe derivada
#Classes e objetos
class Parent:
    #Construtor
     def __init__(self):
          print("Calling parent constructor")

     def __del__(self):
          print("Calling parent destructor")
   
     def parentMethod(self):
          print('Calling parent method')

class Child(Parent):
     def __init__(self):
          print("Calling child constructor")

     def __del__(self):
          print("Calling child destructor")

     def childMethod(self):
          print('Calling child method')

c = Child()
c.childMethod()    
del c  

p = Parent()
p.parentMethod()    
del p 
```

### Bibliotecas

O Python tem várias bibliotecas, ou módulos, que podem ser importados. De cada módulo podem-se também importar submódulos. Ilustra-se o exemplo com as bibliotecas *NumPy*, *SciPy* e *MatPlotLib*.

```python
#Módulos
import numpy as np
import scipy as sp
import matplotlib as mpl
import time

#Submódulos
from matplotlib import pyplot as plt
```

Estas bibliotecas são das mais úteis em Python, e por isso deixam-se alguns exemplos das suas funcionalidades.

- NumPy: significa Numerical Python e é a biblioteca básica de matemática. Contém as funções e operações matemáticas básicas. Uma das suas funcionalidades mais úteis é o *np.array*, que permite criar arrays de números, ou seja, vetores e matrizes.

```python
#NumPy
import numpy as np
print(np.pi + np.sqrt(2))
print(np.exp(np.cos(np.pi)))

v = np.array([0,1])
M = np.array([[0,1],[2,3]])
print(np.matmul(M,v))
```

- SciPy: significa Scientific Python, e é uma extensão do NumPy para ferramentas mais avançadas, em particular métodos numéricos. Tem muitos submódulos, de integração, interpolação, álgebra linear, etc. Como exemplo calcula-se $\int_{0}^{2} \cos{(x)} \ dx$.

```python
#Scipy
import numpy as np
import scipy as sp
v = np.array([0,1])
M = np.array([[0,1],[2,3]])
print(np.matmul(M,v))
print(sp.linalg.det(M))
print(sp.linalg.eigvals(M))

def f(x):
    return np.cos(x)

print(sp.integrate.quad(f, 0, 2))   
```

- *MatPlotLib*: é a biblioteca que permite realizar gráficos matemáticos, sendo muito útil o seu submódulo *PyPlot*. 

```python
#Matplotlib - graphics and plots
import numpy as np
import matplotlib.pyplot as plt

#Criar listas de pontos
x = np.arange(0,100)
y = x**2
fig = plt.figure(figsize=(5, 5))
plt.plot(x , y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Funcao quadratica")
plt.grid()
plt.show()
#plt.savefig('plot.png')
```
